Analyzing data from DataFrames

```python
total_cases = covid_df.Confirmed.sum()
total_deaths = covid_df.Deaths.sum()
```

```python
print("The total no. of reported cases is {}" .format(total_cases))
print(" The total no. of deaths is {}" .format(total_deaths))
```

```python
death_rate = covid_df.Deaths.sum() / covid_df.Confirmed.sum()
```

```python
print("The overall reported death rate is {:.2f}%." .format(death_rate))
```

Working with Dates

```python
covid_df.Date
```

```python
covid_df['Date'] = pd.to_datetime(covid_df.Date)
```

```python
covid_df['Date']
```

```python
covid_df['Year'] = pd.DatetimeIndex(covid_df.Date).year
covid_df['Month'] = pd.DatetimeIndex(covid_df.Date).month
covid_df['Day'] = pd.DatetimeIndex(covid_df.Date).day
covid_df['Weekday'] = pd.DatetimeIndex(covid_df.Date).weekday
```

```python
covid_df
```

```python
covid_df_may = covid_df[covid_df.Month == 5]
covid_df_may
```

```python
covid_df_may_metrics = covid_df_may[['Confirmed', 'Deaths', 'Cured']]
covid_df_may_metrics
```

```python
covid_may_totals = covid_df_may_metrics.sum()
covid_may_totals
```

Grouping and Aggregation

```python
monthly_groups = covid_df.groupby('Month')
monthly_groups
```

```python
monthly_groups[['Confirmed', 'Deaths', 'Cured']].sum()
```

```python
covid_yearly_df = covid_df.groupby('Year')[['Confirmed', 'Deaths', 'Cured']].sum()
```

```python
covid_yearly_df
```

```python
covid_weekday_df = covid_df.groupby('Weekday')[['Confirmed', 'Deaths', 'Cured']].mean()
```

```python
covid_weekday_df
```

```python
covid_df['Total Cases'] = covid_df.Confirmed.cumsum()
```

```python
covid_df['Total Deaths'] = covid_df.Deaths.cumsum()
```

```python
covid_df['Total Cured Cases'] = covid_df.Cured.cumsum()
```
